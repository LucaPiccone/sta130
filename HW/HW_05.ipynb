Now, if the p-value was really low, that means the distributions don't overlap eachother very much.
```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Creating a simple DataFrame for the experiment data
fishers_data = {'Total Tests': [8], 
                 'Correct Predictions': [8], 
                 'Incorrect Predictions': [8 - 8]}

Fishers_df = pd.DataFrame(fishers_data)

# Data for actual predictions
total_tests = Fishers_df['Total Tests'][0]
correct_predictions = Fishers_df['Correct Predictions'][0]
incorrect_predictions = Fishers_df['Incorrect Predictions'][0]

# Proportion of correct predictions
proportion_correct = correct_predictions / total_tests

# Estimate mean and standard deviation for a hypothetical larger sample
mean = proportion_correct * total_tests  # Should be 8.0 (total tests)
std_dev = 1  # Arbitrary standard deviation for visualization

# Generate normal distribution data for actual predictions
x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 1000)
y = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

# Null Hypothesis parameters
p_null = 0.5
mean_null = p_null * total_tests  # Expected mean
std_dev_null = np.sqrt(total_tests * p_null * (1 - p_null))  # Expected standard deviation

# Generate normal distribution data under the null hypothesis
x_null = np.linspace(mean_null - 3 * std_dev_null, mean_null + 3 * std_dev_null, 1000)
y_null = (1 / (std_dev_null * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x_null - mean_null) / std_dev_null) ** 2)

# Plotting both distributions
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Simulated Normal Distribution (Actual Data)', color='blue')
plt.plot(x_null, y_null, label='Null Hypothesis Normal Distribution (p = 0.5)', color='red')
plt.title('Overlapping Normal Distributions')
plt.xlabel('Number of Correct Predictions')
plt.ylabel('Probability Density')
plt.axvline(correct_predictions, color='green', linestyle='dashed', linewidth=1, label='Actual Correct Predictions (8)')
plt.axvline(mean_null, color='purple', linestyle='dashed', linewidth=1, label='Expected Mean (4)')
plt.legend()
plt.grid()
plt.show()

```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

# Creating a simple DataFrame for the experiment data
fishers_data = {'Total Tests': [8], 
                 'Correct Predictions': [8], 
                 'Incorrect Predictions': [8 - 8]}

Fishers_df = pd.DataFrame(fishers_data)

# Given data from the DataFrame
total_tests = Fishers_df['Total Tests'][0]
correct_predictions = Fishers_df['Correct Predictions'][0]
p_null = 0.5  # Null hypothesis proportion

# Sample proportion
p_hat = correct_predictions / total_tests  # Observed proportion

# Calculate the z-score
z = (p_hat - p_null) / np.sqrt((p_null * (1 - p_null)) / total_tests)

# Calculate the p-value (two-tailed)
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

# Output the results
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")
```

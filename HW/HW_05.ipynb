## Population:
The entire group of individuals or objects we're interested in studying. For example, if we want to study the height of men in Canada, the population would be all men in Canada.

## Parameter: (Mu)
A numerical value that describes a characteristic of the population. This value is usually unknown and what we're trying to estimate. In our example, the average height of all men in Canada would be a population parameter.

## Sample: (Xbar)
A subset of individuals or objects selected from the population. We collect data from the sample to make inferences about the population. For instance, we might collect height data from a random sample of 1,000 Canadian men.

## Statistic: 
A numerical value calculated from the sample data. We use statistics to estimate population parameters. The average height of the 1,000 men in our sample would be a sample statistic.

For example, if a statistical test leads us to reject the null hypothesis that "the average height of men in Canada is 5'10"," we're concluding that the population parameter (the average height of all Canadian men) is likely different from 5'10". This conclusion is based on the evidence gathered from our sample, but it applies to the broader population.

The ultimate goal of statistical inference is to gain insights and knowledge about the characteristics of the entire population using the information we glean from a smaller, representative sample.

## conclusion of a statistical test is always about the population parameter, not just the specific sample we've collected data from. 

Statistical tests use sample statistics to assess hypotheses about population parameters. The outcome of a statistical test tells us whether there is enough evidence from the sample data to support or refute a claim about the population.

# 4
a smaller p-value makes the null hypothesis look more ridiculous. Say you have a sample, and from that sample you create a hypothetical distribution to represent the population. Now, you create a new hypothetical distribution of how your data would look if the null hypothesis was used as one of the parameters. Now, overlap the two distributions (one red, and one blue, so you can see the difference of course). If the two distributions overlap eachother alot, the p-value is higher because there is a bigger percentage of the distributions overlapping. This means that your data supports the null hypothesis that you created, the null hypothesis, an idea that you assume about the data. Now, if the p-value was really low, that means the distributions don't overlap eachother very much. If they don't overlap eachother, then you're null hypothesis looks way different from the sample and your null hypothesis can be considered ridiculous. 

A smaller p-value makes the null hypothesis look more ridiculous because it suggests that the observed data is highly unlikely. We have a sample distribution created from the null hypothesis. we have a test statistic. Then, we create simulated samples from the sample distribution the same size as the observed sample. calculate the sample distribution test statistic for each new simulated sample and store it in an array. calculate the distance from each simulations sample distribution test statistic to the null hypothesis. if it is greater than the distance from the observed statistic to the null hypothesis. this would make the p-value larger because your observed statistic is not unusual and provide evidence that you should not reject the null hypothesis.

The higher the p-value the weaker the evidence against the null hypothesis.
More extreme samples from the sampling distribution means higher p-value

```python
import pandas as pd

# Create lists of True/False values
tilt_left = [True] * 44 + [False] * 80
tilt_right = [False] * 44 + [True] * 80

# Create the DataFrame
data = pd.DataFrame({
    "Tilt_left": tilt_left,
    "Tilt_right": tilt_right
})
# Shuffle the rows (optional, if you want to mix up True/False values)
# data = data.sample(frac=1).reset_index(drop=True)

population_parameter_under_H0 = 0.5

observed_statistic = (data.Tilt_right).mean()

print(f'The p = 0.5 Null Hypothesis of 50% right head tilters')
print(f'The observed statistic says {observed_statistic:.2f}% tilt right')
```

```python
import numpy as np

np.random.seed(120)

num_of_simulations = 10000
n = len(data)
simulations_means_under_null_hypothesis_distribution = np.zeros(num_of_simulations)

for i in range(num_of_simulations):
    random_improvement = np.random.choice([0, 1], size=n, replace=True)
    simulations_means_under_null_hypothesis_distribution[i] = random_improvement.mean()
```

```python
simulated_statistic = simulations_means_under_null_hypothesis_distribution

simulated_stats_as_or_more_extreme = \
    abs(simulated_statistic - 0.5) >= abs(observed_statistic - 0.5)

p_value = simulated_stats_as_or_more_extreme.sum() / num_of_simulations
print(p_value)

# This test suggests strong evidence against the null hypothesis.
```

# 6
The p-value can not prove that the null hypothesis is false. Nor can it prove that the null hypothesis is true. It is a measure of of evidence against the null hypothesis.


## The p-value does not represent how likely it is for the null hypothesis to be true 

## Represent the probability of observing data as or more extreme than the data observed based on if the null hypothesis was true.

A low p-value suggests that the observed data is unlikely if the null hypothesis was true.

A high p-value suggests that it likely to get this data if the null hypothesis was true.

Interpret the p-value as a measure of strength against the null hypothesis and not as definitive proof.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

# Creating a simple DataFrame for the experiment data
fishers_data = {'Total Tests': [8], 
                 'Correct Predictions': [8], 
                 'Incorrect Predictions': [8 - 8]}

Fishers_df = pd.DataFrame(fishers_data)

# Given data from the DataFrame
total_tests = Fishers_df['Total Tests'][0]
correct_predictions = Fishers_df['Correct Predictions'][0]
p_null = 0.5  # Null hypothesis proportion

# Sample proportion
p_hat = correct_predictions / total_tests  # Observed proportion

# Calculate the z-score
z = (p_hat - p_null) / np.sqrt((p_null * (1 - p_null)) / total_tests)

# Calculate the p-value (two-tailed)
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

# Output the results
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
import numpy as np
import scipy.stats as stats

# Given data
total_students = 80
correct_predictions = 49
p_null = 0.5

# Sample proportion
p_hat = correct_predictions / total_students

# Calculate the z-score
z = (p_hat - p_null) / np.sqrt((p_null * (1 - p_null)) / total_students)

# Calculate the p-value (two-tailed)
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

# Output the results
print(f"Z-score: {z:.4f}")
print(f"P-value: {p_value:.4f}")
```

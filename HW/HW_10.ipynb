#### Simple Linear Regression versus Multivariate Linear Regression.

Simple linear Regression models the relationship between a single predicator (independent variable) and the outcome (dependent variable). 

The model has the form $Y = B_0 + B_1X + \epsilon$

Where:
- $Y$ = Dependent Variable
- $B_0$ = Intercept (coefficient)
- $B_1$ = Slope (coefficient)
- $X$ = Independent Variable
- $\epsilon$ = Dependent Variable

This assume that $Y$ can be explained with a linear relationship to just one variable X. 

#### Multi Linear Regression models the relationship between multiple predicators (independent variable) to predict the outcome (dependent variable).

The model has the form $Y = B_0 + B_1X_1 + B_2X_2 + ... + B_nX_n + \epsilon$

$X_1, X_2, ...$ = Predicators 

Each predicator has its own coefficient $B_i$ indicating its contribution to $Y$.


#### continuous variable V.S. indicator variable in Simple Linear Regression; and these two linear forms

Continuous variable (age, height, time) 

$Y = B_0 + B_1X + \epsilon$

Indicator variable: a binary variable that take on values of 0 and 1. 

Use an indicator variable when you are interested in comparing two distinct groups in terms of their average outcome.

$Y = B_0 + B_1D + \epsilon$


#### Multi Linear Regression with one continuous variable and One indicator variable.

$Y = B_0 + B_1X + B_2D$

The model can represent two parallel lines, 1 for the D = 1 and another line for when D = 0.

#### Multiple Linear Regression with an interaction term. 

$Y = B_0 + B_1X + B_2D + B_3(X \times D)$

Two - Non parallel lines. 

#### Multiple Linear Regression with Categorical Data.

$Y = B_0 + B_1D_b + B_2D_c$

$D_a$ if $D_b = 0$ and $D_c = 0$

We have a low $R^2$ value of 0.176 but many of the coefficients have a low p-value; supporting evidence against the null hypothesis. This can be explained by the concepts of statistical significance and Explanatory power.

Statistical Significant and Explanatory power are not contradiction of eachother but offer different perspectives of the model. 

Statistical Significance tells us that certain predicators have a meaningful relationship with the outcome variable. Where a high coefficients means the predicator has a large impact on the outcome variable and a low p-value means we can reject the null hypothesis that the predicator has no influence on the outcome. 

On the other hand, Explanatory power, also known as coefficient of determination, measures how well the model captures the variabality in the data. A low $R^2$ value means that a small portion of variability in the data can be explained by the predicators. This can be because important factors are missing or Y is influenced by complex patterns. 

In the example below, Sp Def and Generation have a strong relationship with HP, however, there may be other variables that also affect HP or the relationship between sp def and generation to HP might not be exactly linear.

```python
import numpy as np
from sklearn.model_selection import train_test_split

fifty_fifty_split_size = int(pokeaman.shape[0]*0.5)

# Replace "NaN" (in the "Type 2" column with "None")
pokeaman.fillna('None', inplace=True)

np.random.seed(130)
pokeaman_train,pokeaman_test = \
  train_test_split(pokeaman, train_size=fifty_fifty_split_size)
pokeaman_train
```

```python
model_spec3 = smf.ols(formula='HP ~ Attack + Defense', 
                      data=pokeaman_train)
model3_fit = model_spec3.fit()
model3_fit.summary()
```

```python
yhat_model3 = model3_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model3_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model3)[0,1]**2)
```

```python
model4_linear_form = 'HP ~ Attack * Defense * Speed * Legendary'
model4_linear_form += ' * Q("Sp. Def") * Q("Sp. Atk")'
# DO NOT try adding '* C(Generation) * C(Q("Type 1")) * C(Q("Type 2"))'
# That's 6*18*19 = 6*18*19 possible interaction combinations...
# ...a huge number that will blow up your computer

model4_spec = smf.ols(formula=model4_linear_form, data=pokeaman_train)
model4_fit = model4_spec.fit()
model4_fit.summary()
```

```python
yhat_model4 = model4_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model4_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model4)[0,1]**2)
```

Model 3 and model 4, then model 7 have been scaled and centered around 0. 

#### Model 3 to model 4. 

Model 4 introduces interaction terms between predictors, such as Attack * Defense and Speed * Legendary, to model the more complex relationships between variables and how they jointly influence HP.

The high Condition Number in Model 4 indicates high multicollinearity, suggesting that predictors are highly correlated with each other, leading to potential instability in the estimates.

#### model 4 to model 5

Model 5 extends Model 4 by adding additional predictors like Generation, Type 1, and Type 2. These are categorical variables that provide more information about the Pokémon's characteristics and their potential influence on HP.

#### Model 5 to model 6

Model 6 refines Model 5 by focusing on predictors with the lowest p-values. Binary variables are created for Type 1 = 'Water' and Type 1 = 'Normal', as well as for Generation 2 and Generation 5, because these categories were found to be significant predictors with strong evidence against the null hypothesis (p < 0.05). The goal is to simplify the model by removing predictors that don't significantly contribute to explaining HP.

#### Model 6 to model 7
Model7 adds interaction terms between continuous variables like attack, speed, sp def and sp att. This allows the model to account for more complex relationships between the predictors. If the affects of on predicstor on the outcome relies on other predictors

, this type model will capture those effects more accurately. The inclusion of the centered and scaled version of the model (model7_linear_form_CS) is meant to address potential multicollinearity and ensure that each predictor is on a comparable scale, which can improve model stability and interpretability.

Note we don't center and scale indicator variables.

The large condition number in model7_fit before centering and scaling (2,340,000,000) indicates potential multicollinearity problems—predictors may be highly correlated with each other, leading to unstable estimates. After centering and scaling, the condition number is reduced to a much smaller value (15.4), suggesting that multicollinearity is less of a concern, improving the model’s reliability.

The condition number is a key tool for identifying multicollinearity. 
- Large condition numbers indicate that predictors are highly correlated, which can lead to unstable and unreliable estimates.


The condition number is crucial for detecting multicollinearity. Large values (e.g., greater than 1000) indicate strong correlations between predictors, which can lead to unstable estimates. After centering and scaling, the condition number drops significantly, improving the model's stability and reliability.
